## Exercice 1

#### 1. Création du graphe non dirigé vide


```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
```

#### 2. Ajout des nœuds


```python
G.add_nodes_from(['Mohammed', 'Anas', 'Samir', 'Karim', 'Nizar', 'Amin'])
```

#### 3. Ajout des arêtes

```python
G.add_edges_from([('Mohammed', 'Anas'), ('Mohammed', 'Karim'), ('Anas', 'Samir'), ('Amin', 'Nizar')])
```

#### 4. Affichage du graphe


```python
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=800, node_color='skyblue', font_color='black', edge_color='gray')
plt.title("Réseau Social")
plt.show()
```

#### 5. Nombre total de nœuds


```python

total_nodes = G.number_of_nodes()
print(f"Nombre total de nœuds dans le graphe : {total_nodes}")
```

#### 6. Nombre total d'arêtes


```python
total_edges = G.number_of_edges()
print(f"Nombre total d'arêtes dans le graphe : {total_edges}")
```

#### 7. Liste des voisins de chaque nœud


```python
neighbors_list = {node: list(G.neighbors(node)) for node in G.nodes}
print("Liste des voisins de chaque nœud :")
for node, neighbors in neighbors_list.items():
    print(f"{node}: {neighbors}")
```

#### 8. Vérification de l'amitié entre Mohammed et Nizar


```python

are_friends = G.has_edge('Mohammed', 'Nizar')
print("Mohammed et Nizar sont-ils amis ? Réponse :", are_friends)
```

#### 9. Suppression de l'amitié entre Anas et Samir


```python
G.remove_edge('Anas', 'Samir')
```

#### 10. Nombre total de nœuds et d'arêtes après suppression


```python

new_total_nodes = G.number_of_nodes()
new_total_edges = G.number_of_edges()
print(f"\nNombre total de nœuds après suppression : {new_total_nodes}")
print(f"Nombre total d'arêtes après suppression : {new_total_edges}")
```

#### 11. Liste des voisins de chaque nœud après suppression de l'arête


```python
new_neighbors_list = {node: list(G.neighbors(node)) for node in G.nodes}
print("\nListe des voisins de chaque nœud après suppression de l'arête :")
for node, neighbors in new_neighbors_list.items():
    print(f"{node}: {neighbors}")
```

## Exercice 3

#### 1. Création du graphe non dirigé

```python
import networkx as nx
import matplotlib.pyplot as plt
import nxviz as nv

G = nx.Graph()
```

#### Ajout des personnes et des connexions d'amitié (exemple)


```python
people = ['Pers1', 'Pers2', 'Pers3', 'Pers4', 'Pers5', 'Pers6', 'Pers7', 'Pers8']
edges = [('Pers1', 'Pers2'), ('Pers1', 'Pers3'), ('Pers1', 'Pers4'),
         ('Pers2', 'Pers5'), ('Pers2', 'Pers6'), ('Pers3', 'Pers7'),
         ('Pers4', 'Pers8'), ('Pers5', 'Pers6'), ('Pers6', 'Pers7')]

G.add_nodes_from(people)
G.add_edges_from(edges)
```

#### 2. Affichage du graphe


```python

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=800, node_color='skyblue', font_color='black', edge_color='gray')
plt.title("Réseau Social")
plt.show()
```

#### 3. Visualisation avec Nxviz

```python
arc_plot = nv.ArcPlot(G)

plt.title("Visualisation avec Nxviz")
plt.show()
```

#### 4. Recherche de chemins simples entre deux personnes


```python

person_a, person_b = 'Pers1', 'Pers7'
paths = list(nx.all_simple_paths(G, source=person_a, target=person_b))
print(f"Chemins entre {person_a} et {person_b} : {paths}")
```

#### 5. Calcul des degrés de centralité


```python

degree_centrality = nx.degree_centrality(G)
print("\nDegrés de centralité :")
for person, centrality in degree_centrality.items():
    print(f"{person}: {centrality}")
```

#### 6. Calcul de l'intermédiarité


```python
betweenness_centrality_ex3 = nx.betweenness_centrality(G)
print("\nIntermédiarité (Betweenness Centrality) :")
for person, centrality in betweenness_centrality_ex3.items():
    print(f"{person}: {centrality}")
```

#### 7. Identification des cliques


```python
cliques = list(nx.find_cliques(G))
print("\nCliques identifiées : ", cliques)
```

#### 8. Création d'un sous-graphe à partir d'une clique


```python
#['Pers5', 'Pers2', 'Pers6']on vas utiliser cette clique
clique_to_include = ['Pers5', 'Pers2', 'Pers6']
subgraph = G.subgraph(clique_to_include)
```

#### 9. Affichage du sous-graphe


```python

pos_subgraph = nx.spring_layout(subgraph)
nx.draw(subgraph, pos_subgraph, with_labels=True, font_weight='bold', node_size=600, node_color='lightcoral', font_color='black', edge_color='darkred')
plt.title("Sous-graphe d'une Clique Identifiée")
plt.show()
```
